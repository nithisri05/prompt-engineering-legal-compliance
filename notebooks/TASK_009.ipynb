Use Case 3:
  * Description : Identify early signs of emerging policy drift before major non-compliance occurs
  * Prompt Template : Analyze the provided behavior patterns and identify early warning indicators of potential policy drift.


```python
import openai

client = openai.OpenAI(
    api_key="sk-QK5vbrlNoAVpUMqaeWzMDw",     # paste your key
    base_url="https://apidev.navigatelabsai.com/"
)
```

```python
def run_llm(prompt):
    """
    Sends a prompt to the LLM and returns the response.
    """
    response = openai.ChatCompletion.create(
        model="gpt-4.1-nano",
        messages=[
            {"role": "system", "content": "You are an expert in Governance, Risk, and Compliance."},
            {"role": "user", "content": prompt}
        ],
        temperature=0.3
    )
    return response.choices[0].message.content
```

```python
#Prompt for generating output
prompt_uc9 = """
Analyze the behavior trends and identify early warning
indicators of emerging policy drift.
"""
```

```python
#Generates the Output based on the given prompt
output = run_llm(prompt_uc9)
print(output)
```

  Sample Input & Output
  * Input:
An audit and compliance prompt asking the LLM to analyze observed behavior trends and identify early warning indicators of emerging policy drift, such as deviations from security or access control policies over time.


* Output:
A textual analysis where the LLM highlights signs like gradual increase in policy exceptions, late or skipped approvals, lax password or access practices, and rising minor violations as early indicators that policy drift is beginning to emerge
