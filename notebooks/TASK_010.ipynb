 Use Case 1:
 * Description : Simulate regulatory evaluation of organizational actions or incidents
 * Prompt Template : Act as a regulatory authority and assess the following scenario for compliance risks and violations.


```python
import openai

client = openai.OpenAI(
    api_key="sk-QK5vbrlNoAVpUMqaeWzMDw",     # paste your key
    base_url="https://apidev.navigatelabsai.com/"
)
```

```python
def run_llm(prompt):
    """
    Sends a prompt to the LLM and returns the response.
    """
    response = client.chat.completions.create(
        model="gpt-4.1-nano",
        messages=[
            {"role": "system", "content": "You are an expert in Governance, Risk, and Compliance."},
            {"role": "user", "content": prompt}
        ],
        temperature=0.3
    )
    return response.choices[0].message.content
```

```python
#Prompt for generating output
prompt_uc10 = """
Act as a regulatory authority.
Assess the following scenario for compliance gaps:
Delayed breach notification beyond regulatory timelines.
"""
```

```python
#Generates the Output based on the given prompt
output = run_llm(prompt_uc10)
print(output)
```

Sample Input & Output

* Input:
A regulatory‑focused prompt instructing the LLM to act as a regulatory authority and assess a scenario where a breach notification was sent later than required by law, asking it to identify resulting compliance gaps and implications.
​

* Output:
A textual compliance assessment describing that delayed breach notification likely violates regulatory timelines, increases legal and enforcement risk, may trigger penalties or supervisory actions, and indicates weaknesses in incident response and notification procedures.

Use Case 3:
* Description : Generate regulator-style recommendations for corrective actions.
* Prompt Template : As a regulator, provide recommendations and corrective actions required to address the identified compliance issues.


```python
from openai import OpenAI
from IPython.display import Audio, display


client = OpenAI(
    api_key="sk-QK5vbrlNoAVpUMqaeWzMDw", # paste your key
    base_url="https://apidev.navigatelabsai.com"
)

audio_bytes = b""

# Regulator-style compliance recommendation (TEXT → SPEECH)
regulator_text = """
From a regulatory perspective, the organization has failed to meet
mandatory breach notification timelines.

Immediate corrective actions are required, including updating incident
response procedures, conducting staff training, and implementing
automated alert mechanisms.

Failure to comply may result in enforcement action and financial penalties.
"""

# Stream text-to-speech response
with client.audio.speech.with_streaming_response.create(
    model="gpt-4o-mini-tts",
    voice="alloy",
    input=regulator_text
) as response:
    for chunk in response.iter_bytes():
        audio_bytes += chunk

# Play audio in Google Colab
display(Audio(audio_bytes, autoplay=True))
```

Sample Input & Output

* Input:
A regulatory statement explaining that the organization missed mandatory breach notification timelines, requires immediate corrective actions such as updating incident response procedures, staff training, and automated alerts, and warning that continued non‑compliance may lead to enforcement actions and financial penalties.
​

* Output:
A streamed text‑to‑speech audio clip that clearly narrates the regulator’s findings, the required remediation steps, and the potential consequences, which is then played back automatically in the notebook using the generated audio bytes
